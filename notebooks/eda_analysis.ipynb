```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.barplot(x='COUNTRY', y='SALES', data=df)
plt.xticks(rotation=90)
plt.title('Sales by Country')

plt.savefig("../images/sales_by_country.png")

plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='PRODUCTLINE', y='SALES', data=df)
plt.xticks(rotation=45)
plt.title('Sales by Product Line')

plt.savefig("../images/productline_sales.png")

plt.show()
```

```python
sns.countplot(x='DEALSIZE', data=df)

plt.savefig("../images/dealsize.png")

plt.show()
```
